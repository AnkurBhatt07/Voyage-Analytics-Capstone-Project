```python
import numpy as np
import pandas as pd 
```

```python
data_path = "gender_data/agg_user_data.csv"

df = pd.read_csv(data_path)
```

```python
# dropping the userCode
df = df.drop(columns=['userCode'])
```

```python
X = df.drop(columns=['gender'])
y = df['gender']
```

```python
import json 
idx = np.random.randint(0,len(df))
print(f"The row index is : {idx}")
sample_row = X.iloc[idx]

input_dict = sample_row.to_dict()

input_dict = {k:(v.item() if hasattr(v,'item') else v) for k , v in input_dict.items()}

json_input = json.dumps(input_dict , indent = 2)

print(json_input)
```
