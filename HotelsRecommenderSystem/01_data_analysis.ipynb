```python
import numpy as np 
import pandas as pd 


hotels_df = pd.read_csv('../data/hotels.csv')
hotels_df.shape
```

```python
interactions = (
    hotels_df.groupby(['userCode','name','place']).size().reset_index(name = 'interaction')   
)

interactions.head()
```

```python
print("Users:", interactions['userCode'].nunique())
print("Hotels" , interactions['name'].nunique())
print("Interactions:", len(interactions))
```

```python
# Popular Hotels per place 

popular_by_place = (
    interactions.groupby(['place','name']).agg(total_interactions = ('interaction' , 'sum')).reset_index().sort_values(['place','total_interactions'] , ascending = [True , False])
)

popular_by_place.head(10)
```

```python
all_places = popular_by_place['place'].nunique()
# Chose one place randomly
rand_place = popular_by_place['place'].sample(1).values[0]
print(f"Total Places: {all_places} , Random Place: {rand_place}")
```

```python
# Recommend the best hotel for the random place using popular_by_place df 
popular_by_place.loc[popular_by_place['place'] == rand_place, 'name' ].values[0]
```
